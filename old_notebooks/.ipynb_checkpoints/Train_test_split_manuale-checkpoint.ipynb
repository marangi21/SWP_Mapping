```python
import pandas as pd
```

```python
#dataset
input_path_1 = r"C:\Users\giova\OneDrive\Desktop\ZEULI\3 Zeuli 220623\3 Zeuli 220623 ground truth.xlsx"
input_path_2 = r"C:\Users\giova\OneDrive\Desktop\ZEULI\5 Zeuli 220802\5 Zeuli 220802 ground truth.xlsx"

#alberi selezionati per far parte del test set
alberi_validation_set = [51, 52, 53, 54, 148]
alberi_test_set = [59, 60, 141, 142, 146, 147]

df_1 = pd.read_excel(input_path_1)
df_2 = pd.read_excel(input_path_2)
```

```python
concat_df = pd.concat([df_1, df_2], ignore_index=True)
concat_df.insert(0, 'ID', concat_df.reset_index().index+1)
concat_df
```

```python
concat_df.to_csv(r'C:\Users\giova\OneDrive\Desktop\ZEULI\dataset\dataset.csv')
```

```python
#estrazione del test set
df_test = concat_df[concat_df['ID_Albero'].isin(alberi_test_set)]
print(f"Valori unici ID_Albero in df_test = {df_test['ID_Albero'].unique()}")
df_test
```

```python
#estrazione del validation set
df_val = concat_df[concat_df['ID_Albero'].isin(alberi_validation_set)]
print(f"Valori unici ID_Albero in df_val = {df_val['ID_Albero'].unique()}")
df_val
```

```python
#estrazione del training set
df_train = concat_df.drop(df_test.index)
df_train = df_train.drop(df_val.index)
df_train
```

```python
#check se training set e test set hanno record in comune
df_common = df_train.isin(df_test)
df_common['ID_Pixel'].unique()
```

```python
#check se training set e validation set hanno record in comune
df_common = df_train.isin(df_val)
df_common['ID_Pixel'].unique()
```

```python
#check se training set e validation set hanno record in comune
df_common = df_test.isin(df_val)
df_common['ID_Pixel'].unique()
```

```python
df_train.to_csv(r'C:\Users\giova\OneDrive\Desktop\ZEULI\dataset\training_set.csv')
df_val.to_csv(r'C:\Users\giova\OneDrive\Desktop\ZEULI\dataset\validation_set.csv')
df_test.to_csv(r'C:\Users\giova\OneDrive\Desktop\ZEULI\dataset\test_set.csv')
```
