```python
from osgeo import gdal
import sentinelhub as sh
from sentinelhub import CRS
import rasterio
import pandas as pd
import matplotlib.pyplot as plt
```

```python
input_folder = r"C:\Users\giova\OneDrive\Desktop\ZEULI"
input_path = input_folder + "\\UNICO_ZEULI_OLIVO.xlsx"
output_path = input_folder + "\\ZEULI_OLIVO_UTM_220623.xlsx"
data_volo = "2022-06-23"
```

```python
#Apre foglio excel in un dataframe
df = pd.read_excel(input_path, header=1)
#print(df)
df_day = df[df["DATA  RILIEVO"] == data_volo]
#print(df[["LONGITUDINE", "LATITUDINE"]])
print(df_day[["LONGITUDINE", "LATITUDINE", "ID"]])
```

```python
#CRS delle immagini
target_crs = CRS.UTM_33N
#CRS standard lon/lat
source_crs = CRS.WGS84
```

```python
# conversione punti da lon/lat in source_crs a UTM in target_crs
utm_coords = []
for lon, lat, id in zip(df_day['LONGITUDINE'], df_day['LATITUDINE'], df_day['ID']):
    east, north = sh.transform_point((lon, lat), source_crs, target_crs)
    utm_coords.append((east,north,id))

utm_coords
utm_df = pd.DataFrame(utm_coords, columns = ['UTM_lon', 'UTM_lat', 'ID'])
utm_df
```

```python
df_day = pd.merge(df_day, utm_df, on='ID')
df_day
```

```python
# Seleziona solo le colonne desiderate nell'ordine specificato
df_new = df_day[['DATA  RILIEVO', 'ID', 'LONGITUDINE', 'LATITUDINE', 'UTM_lon', 'UTM_lat', 'TEMP ARIA', 'UMID ARIA', 'SWP']]

df_new.loc[:, 'TEMP ARIA'] = pd.to_numeric(df_new['TEMP ARIA'], errors='coerce')
df_new = df_new.dropna(subset=['TEMP ARIA'])
df_new.loc[:, 'UMID ARIA'] = pd.to_numeric(df_new['UMID ARIA'], errors='coerce')
df_new = df_new.dropna(subset=['UMID ARIA'])
df_new.loc[:, 'SWP'] = pd.to_numeric(df_new['SWP'], errors='coerce')
df_new = df_new.dropna(subset=['SWP'])

df_new['DATA  RILIEVO'] = pd.to_datetime(df_new['DATA  RILIEVO'])
df_new['DATA  RILIEVO'] = df_new['DATA  RILIEVO'].dt.date

# Stampa del nuovo DataFrame
print(df_new)
```

```python
df_new.to_excel(output_path, index=False)
```
