```python
from osgeo import gdal, gdal_array
import numpy as np
import rasterio as rio
import matplotlib.pyplot as plt
import spyndex as spy
import math
import xarray as xr
from rasterio import plot
```

```python
# File di input e output
input_tif = r"C:\Users\giova\OneDrive\Desktop\Resampled\1 Tormaresca 220526\MULTI_ORTHO\1 Tormaresca 220526 ORTHO_MULTI_10.tif"
output_tif = r"C:\Users\giova\OneDrive\Desktop\Resampled\1 Tormaresca 220526\MULTI_ORTHO\1 Tormaresca 220526 ORTHO_MULTI_R_G_B_NIR_RE"

# Bande da estrarre
band_indices = [1, 3, 5, 7, 9]

# Apre il raster di input
input_ds = gdal.Open(input_tif)

# Ottiene informazioni sulla larghezza, altezza e numero di bande del raster di input
width = input_ds.RasterXSize
height = input_ds.RasterYSize
num_bands = len(band_indices)

# Ottiene il tipo di dati del raster di input
data_type = input_ds.GetRasterBand(1).DataType

# Crea il raster di output
driver = gdal.GetDriverByName('GTiff')
output_ds = driver.Create(output_tif, width, height, num_bands, data_type)

# Copia le informazioni di geotrasformazione e sistema di riferimento dal raster di input a quello di output
output_ds.SetGeoTransform(input_ds.GetGeoTransform())
output_ds.SetProjection(input_ds.GetProjection())

# Estrae e copia le bande specificate
for i, band_index in enumerate(band_indices, start=1):
    band_data = input_ds.GetRasterBand(band_index).ReadAsArray()
    output_ds.GetRasterBand(i).WriteArray(band_data)

# Chiude i dataset
input_ds = None
output_ds = None

print("Processo completato. Il nuovo raster è stato creato:", output_tif)
```

```python
with rio.open(r"C:\Users\giova\OneDrive\Desktop\Resampled\1 Tormaresca 220526\MULTI_ORTHO\1 Tormaresca 220526 ORTHO_MULTI_R_G_B_NIR_RE") as src:
    b1 = src.read(1)
    b2 = src.read(2)
    b3 = src.read(3)
    b4 = src.read(4)
    b5 = src.read(5)

    red = b1.astype('float64')
    green = b2.astype('float64')
    blue = b3.astype('float64')
    nir = b4.astype('float64')
    rededge = b5.astype('float64')

    data = xr.DataArray(
    np.array([red, green, blue, nir, rededge]).reshape(5,src.width,src.height),
    dims = ("band","x","y"),
    coords = {"band": ["Red", "Green", "Blue", "NIR", "Rededge-705"]}
    )

    idx = spy.computeIndex(
        index = ['NDVI', 'OSAVI', 'TCARI', 'TCARIOSAVI', 'EVI', 'MCARI', 'MCARIOSAVI', 'SR', 'TVI', 'NDREI'],
        params = {
            "R": data.sel(band = "Red"),
            "G": data.sel(band = "Green"),
            "B": data.sel(band = "Blue"),
            "N": data.sel(band = "NIR"),
            "RE1": data.sel(band = "Rededge-705"),
            "g": spy.constants.g.default,
            "C1": spy.constants.C1.default,
            "C2": spy.constants.C2.default,
            "L": spy.constants.L.default
        }
    )
```

```python
fig, ax = plt.subplots(2, 2, figsize = (10,10))
plot.show(data.sel(band = ["Red", "Green", "Blue"]).data/0.3, ax=ax[0,0], title = "RGB")
plot.show(idx.sel(index = "NDVI").data, ax=ax[0,1], title="NDVI")
plot.show(idx.sel(index = "TCARI").data, ax=ax[0,1], title="TCARI")
plot.show(idx.sel(index = "NDREI").data, ax=ax[0,1], title="NDREI")
```
