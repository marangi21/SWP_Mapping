```python
# conversione punti da lon/lat in source_crs a UTM in target_crs
utm_coords = []
for lon, lat, id in zip(df_day['LONGITUDINE'], df_day['LATITUDINE'], df_day['ID']):
    east, north = sh.transform_point((lon, lat), source_crs, target_crs)
    utm_coords.append((east,north,id))

utm_coords
utm_df = pd.DataFrame(utm_coords, columns = ['UTM_lon', 'UTM_lat', 'ID'])
utm_df
```

```python
df_day = pd.merge(df_day, utm_df, on='ID')
df_day
```
